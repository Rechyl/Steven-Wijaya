```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# Load data
from google.colab import drive

drive.mount('/content/drive')
directory= '/content/drive/MyDrive/survey lung cancer.csv'
sw = pd.read_csv(directory)
sw.head()
```

```python
sw.columns = ["GENDER", "AGE", "SMOKING", "YELLOW FINGERS", 
              "ANXIETY", "PEER PRESSURE", "CHRONIC DISEASE", 
              "FATIGUE", "ALLERGY", "WHEEZING", "ALCOHOL CONSUMING", 
              "COUGHING", "SHORTNESS OF BREATH", "SWALLOWING DIFFICULTY", 
              "CHEST PAIN", "LUNG CANCER"]
```

```python
import matplotlib.pyplot as plt
# plots other feature sets
sw.hist(figsize = (20,20))
plt.show()
```

```python
# transforming dataset with dummies variables to replace characters with binary integers
sw['GENDER'].replace(["M","F"], [0,1], inplace=True)
sw['LUNG CANCER'].replace(["YES","NO"],[0,1], inplace=True)
```

```python
sw.head()
```

```python
#feature set
x = sw[["GENDER", "AGE", "SMOKING", "YELLOW FINGERS", "ANXIETY", "PEER PRESSURE", "CHRONIC DISEASE","FATIGUE", "ALLERGY", "WHEEZING", "ALCOHOL CONSUMING", "COUGHING", "SHORTNESS OF BREATH", "SWALLOWING DIFFICULTY", "CHEST PAIN"]]

#label set
y = sw["LUNG CANCER"]
```

```python
#train test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

print("Number transactions x_train dataset: ", x_train.shape)
print("Number transactions y_train dataset: ", y_train.shape)
print("Number transactions x_test dataset: ", x_test.shape)
print("Number transactions y_test dataset: ", y_test.shape)
```

```python
# Train Model
model = GaussianNB()
model.fit(x_train, y_train)
```

```python
#Test Model
y_pred = model.predict(x_test)
print('Accuracy : ', accuracy_score(y_test, y_pred))
```
